```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
df = pd.read_csv('dataset/Salary_Data.csv')
```

```python
'''
Penggunaan info
untuk melihat informasi data frame, seperti jumlah baris, 
kolom, nilai non-NULL, tipe data, dan total penggunaan memori.
'''
df.info()
```

```python
# Visualisasi korelasi dengan menggunakan heatmap.
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))

sns.heatmap(df.corr('pearson'), annot=True, cmap='YlGnBu', linewidths=0.2)
```

```python
# Histogram dapat membantu dalam melihat persebaran data.
sns.histplot(data = df, x= 'Salary' ,color='orange', kde=False)
```

```python
# Bar chart digunakan untuk membandingkan data-data yang berbentuk kategorikal.
df2 = pd.read_csv('dataset/Mall_Customers.csv')
sns.barplot(x='Age',y='Annual Income',data = df2, hue='Gender', errcolor='teal')
```

```python
# Pie chart membantu memvisualisasikan komposisi elemen-elemen dalam sebuah data.
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.legend(y, loc='upper left')
plt.show() 
```
